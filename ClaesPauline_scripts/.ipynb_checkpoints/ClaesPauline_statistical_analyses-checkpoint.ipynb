## Hypotheses
1. We will observe the following differences in the _go_-constructions in general:

    a) Across both studied time frames, (`early` and `later`), both _go_-constructions together (progressive [BE Ving]+[GO to INF] and non-progressive [GO to V]) will display a higher normalized frequency in translations than in reference texts, indicating that the constructions are generally used more in translations than in reference texts, which would point toward the possibility of their use being influenced by the presence of _futur proche_ in the source texts. 
    
    b) Across both text types (translations and reference texts), both _go_-constructions together (progressive [BE Ving]+[GO to INF] and non-progressive [GO to V]) will display a higher normalized frequency in `later` than in `early`, indicating that these constructions have become more conventional over time. 
    
    c) The French source text corpus will display a significantly higher normalized frequency of the _go_-construction than the English corpus (translation and reference corpus together) in `early`, but this difference will no longer be significant in `later`. This would confirm that the French _go_-construction was already more conventional at that time, and that the English _go_-construction increased in frequency accordingly.
    
    d) Both _go_-constructions together (progressive and non-progressive) will display a significantly higher normalized frequency in reference texts in `later` than in `early`, indicating that both constructions have become more conventional in non-translated language over time. However, the difference in normalized frequency of the _go_-construction in French source texts between `early` and `later` will not be significant, as the construction had already conventionalized earlier than in English.
    
    e) The progressive [\textsc{be} Ving]+[\textsc{go} to INF] and non-progressive [\textsc{go} to V] constructions separately will display a higher normalized frequency in reference texts in `later` as compared to `early`, indicating that each of the constructions individually have become more conventional in non-translated language over time.
    
    
2. Regarding progressive [BE Ving]+[GO to INF] and non-progressive [GO to V] separately, the `early` translations will display the following differences from the `early` reference texts, measured in proportions as well as in normalized frequencies:

    a) The translations will use the _go_-construction more often in non-motion contexts than the reference texts. 
    
    b) The translations will use the _go_-construction more often in contexts with non-agentive or inanimate subjects than the reference texts. 
    
    c) The translations will use the _go_-construction more often in contexts where there is no intentionality expressed than the reference texts. 
    
    d) The translations will use the _go_-construction more often in contexts about pure predictions than the reference texts. 
    
    e) The translations will use the construction more frequently in a passive voice than the reference texts. 
    
    
3. The same trends as in (2) will be observable in `later` but the discrepancy in normalized frequencies between the translations and reference texts will be smaller than in `early`, indicating that the use of these constructions in these contexts has increased in `later` compared to `early`, and therefore, that the construction has become more conventional in non-translated language over time.


4. The same trends as in (2) will be observable in the `early` source text corpus as compared to the `later` source text corpus. 


5. Using binary logistic mixed effects regression, the above discussed parameters (_motion, animacy, intentionality, predictiveness,_ and _voice_) as well as the use of progressive or non-progressive _go_-constructions will be significant predictors for whether or not a text is a translation or a reference text in `early`, but these will no longer be significant predictors in `later`, indicating that the construction has become more conventional in non-translated language over time.

### 1A.
Across both studied time frames, (`early` and `later`), both _go_-constructions together (progressive [BE Ving]+[GO to INF] and non-progressive [GO to V]) will display a higher normalized frequency in translations than in reference texts, indicating that the constructions are generally used more in translations than in reference texts, which would point toward the possibility of their use being influenced by the presence of _futur proche_ in the source texts. 

```r
unique(meta$transl_ref_srcTxt)
```

```r
# wordcount of all translations (progressive AND nonProgressive, early AND later)
translation_wc <- sum(subset(meta, transl_ref_srcTxt=="transl")$wordcount)
# wordcount of all reference texts (progressive AND nonProgressive, early AND later)
reference_wc <- sum(subset(meta, transl_ref_srcTxt=="ref")$wordcount)
```

```r
# calculating normalized frequencies
translation_relfreq <- (nrow(filter(en, kind=="translation")) / translation_wc)*fixedNr
reference_relfreq <- (nrow(filter(en, kind=="reference")) / reference_wc)*fixedNr
```

```r
transl_ref_relfreq <- matrix(c(translation_relfreq, reference_relfreq), 
                            ncol=2, nrow=1)
colnames(transl_ref_relfreq) <- c('translation', 'reference')
rownames(transl_ref_relfreq) <- "rel.freq."

transl_ref_relfreq
```

> The relative frequency of both _go_-constructions is indeed higher in translations than in reference texts, across both time frames. 

> Now, we will see if this difference in relative frequencies is significant (if there exists a significant positive correlation between the normalized frequency and the subcorpus). Because correlation tests need numerical data, 'translation' was coded as 1, and 'reference' was coded as 0. In addition, normalized frequencies were calculated for each text for both _go_-constructions taken together, as well as each of the _go_-constructions separately. This concretely means that for each text, the normalized frequency (pmw) of the construction(s) in question <u>in that text</u> was calculated. A correlation test (Kendall's rank correlation coefficient $\tau$) will then measure whether the normalized frequencies of a certain 'kind' (translation or reference) are significantly higher/lower.

> For this, we need the `en_textStats` and `fr_textStats` data frames.

```r
head(en_textStats)
```

This data frame contains the text statistics per text for each of the researched variables. For more information, please read the README.md file. As already discussed, we need numerical data for the correlation tests. Because we want to measure possible correlations between the use of constructions and the time frame or the kind of texts, these were made into binary variables:

- `timeframe`:
    - 0 = `early`
    - 1 = `later`
- `kind`:
    - 0 = `reference`
    - 1 = `translation`


In addition, if we want to measure a possible correlation between the use of constructions and the period (e.g. 1580-1589) of the text in which they occur, we also need a numerical variable rather than a categorical one. That is why each period was assigned a number chronologically. This means that the first studied period, 1580-1589, is assigned 1, and the last period, 1690-1699, is assigned 5:

- `period_category`:
    - 1: 1580-1589
    - 2: 1590-1599
    - 3: 1600-1609
    - 4: 1680-1689
    - 5: 1690-1699

```r
# casting 'kind' (translation vs. reference) to a factor for the boxplot that we will use. 
en_textStats$kind <- as.factor(en_textStats$kind)

# checking the levels of this factor, to see which is the reference level.
levels(en_textStats$kind)
```

> 'reference' is the reference level, which is in accordance with my binary encoding (0 = reference, 1 = translation)

```r
boxplot(norm_both~kind, data=en_textStats,
            # norm_both is the normalized frequency of both go-constructions per text
        names=c('reference', 'translation'),
        xlab='Subcorpus (kind)',
        ylab='Normalized frequency (per million words)',
        main='Normalized frequency (pmw) of both \n go-constructions in translations vs. reference texts \n across both time frames'
    )
```

```r
# Kendall's tau rank correlation coefficient
# alternative = greater because we expect a positive correlation: we expect that translations (1) will have 
# a significantly higher normalized frequency than reference texts. 
cor.test(en_textStats$kind_binary, en_textStats$norm_both, method='kendall', alternative='greater')
```

> There seems to be a very slight positive correlation, but not significant.

> If we want to do a wilcoxon test on samples from two different data frames, we need to create new data frames.

## Hypothesis 5.

Binary Logistic Mixed Effects Regression. Using binary logistic mixed effects regression, the above discussed parameters (_motion, animacy, intentionality, predictiveness,_ and _voice_) as well as the use of progressive or non-progressive _go_-constructions will be significant predictors for whether or not a text is a translation or a reference text in `early`, but these will no longer be significant predictors in `later`, indicating that the construction has become more conventional in non-translated language over time. 

```r
en$author_id <- as.character(en$author_id)
```

```r
en_early <- filter(en, timeframe=='early')
en_later <- filter(en, timeframe=='later')
```

### Predicting KIND

### `--- early ---`

```r
en_early <- filter(en_early, motion != 'indeterminate' & intentionality != 'amb')
```

```r
en_early$kind <- as.factor(en_early$kind)
en_early$aspect <- as.factor(en_early$aspect)
en_early$motion <- as.factor(en_early$motion)
en_early$anim_binary <- as.factor(en_early$anim_binary)
en_early$intentionality <- as.factor(en_early$intentionality)
en_early$predictiveness_binary <- as.factor(en_early$predictiveness_binary)
en_early$voice <- as.factor(en_early$voice)
```

```r
unique(en_early$kind)
unique(en_early$aspect)
unique(en_early$motion)
unique(en_early$anim_binary) # only one value
unique(en_early$intentionality) 
unique(en_early$predictiveness_binary) 
unique(en_early$voice)
```

```r
model1 <- glmer(kind ~ aspect
                + motion 
          #      + anim_binary  # only has 1 value.: animate
                + intentionality 
                + predictiveness_binary 
                + voice 
                + (1|author_id),
               data = en_early,
               family = binomial)

drop1(model1, test='Chisq')
```
